#Importação dos dados





```python
url='https://raw.githubusercontent.com/PDI-2022/Testes/main/Dados%20Tetraz%C3%B3lio%20Feij%C3%A3o%20-%20Plan1%20(1).csv'
sementes = pd.read_csv(url,sep=",")
```

#Agrupando os resultados obtidos em Classes (Classe Experimental)






```python
sementes ['CLASSES EXP']= sementes['% VERMELHO']
sementes['CLASSES EXP'] = np.where(  (sementes['CLASSES EXP'] < 10.00), 5, sementes['CLASSES EXP'])
sementes['CLASSES EXP'] = np.where( (sementes['CLASSES EXP'] >= 10.00) & (sementes['CLASSES EXP'] < 40.00), 4, sementes['CLASSES EXP'])
sementes['CLASSES EXP'] = np.where( (sementes['CLASSES EXP'] >= 40.00) & (sementes['CLASSES EXP'] < 60.00), 3, sementes['CLASSES EXP'])
sementes['CLASSES EXP'] = np.where( (sementes['CLASSES EXP'] >= 60.00) & (sementes['CLASSES EXP'] < 80.00), 2, sementes['CLASSES EXP'])
sementes['CLASSES EXP'] = np.where( sementes['CLASSES EXP'] >= 80.00, 1, sementes['CLASSES EXP'])
sementes
```

#Comparando as classes teóricas com as obtidas experimentalmente

```python
sementes['COMP'] = sementes['CLASSE'] - sementes['CLASSES EXP']
sementes.loc[sementes['COMP'] < 0, 'COMP'] = -1
sementes.loc[sementes['COMP'] > 0, 'COMP'] = 1
sementes.loc[sementes['COMP'] == 0, 'COMP'] = 0
sementes['COMP'].replace({-1:'Modelo é maior', 1:'Modelo é menor', 0:'Iguais'}, inplace=True)
sementes
```
